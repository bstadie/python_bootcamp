Find all of the job titles and all of the slaries. 

```python
import csv

# Open the CSV file and create a CSV reader object with DictReader
with open('ds_salaries.csv', 'r') as file:
    csv_reader = csv.DictReader(file)
    all_salaries = []
    all_job_titles = []
    # Iterate over the rows
    for row in csv_reader:
        # Access column values using the header keys
        salary= row['salary_in_usd']
        salary = float(salary)
        job_title = row['job_title']
        all_salaries.append(salary)
        all_job_titles.append(job_title)
        # ...
        
        # Process the row as needed
        # print(column1, column2)

unique_elements = set(all_job_titles)
print(F"There are {len(unique_elements)} unique jobs.")
#print(F"Unique Jobs: {unique_elements}")
import pprint
pprint.pprint(unique_elements)
print(all_salaries)
```

We want to compute the average salary across each job title. Our idea is to keep a dctionary that has a running sum of the total money made at that job 

Then another dictionary of the total number of people that work that job. 

We can compute the mean at the end.



```python
salary_running_sum = {}
salary_count_per_job = {}


for salary, title in zip(all_salaries, all_job_titles):
    if title not in salary_running_sum:
        salary_running_sum[title] = salary
        salary_count_per_job[title] = 1
    else:
        salary_running_sum[title] += salary
        salary_count_per_job[title] += 1
        #print(F"Salary for {title}: {salary_running_sum[title] / salary_count_per_job[title]}")

average_salary = {}
print(salary_running_sum)
print(salary_count_per_job)
for title in salary_running_sum:
    average_salary[title] = salary_running_sum[title] / salary_count_per_job[title]

print("Average Salary per Job Title:")
for title, average in average_salary.items():
    print(title, ":", average)

print(F"Average salary 3d researcher: {average_salary['3D Computer Vision Researcher']}")
```

What if we used a pandas dataframe instead of loading things in manually. 

```python

import pandas as pd


# Read the CSV file
df = pd.read_csv('ds_salaries.csv')

# Display the first few rows of the DataFrame
print(df.head())
```

Use pandas to compute the average salary for each job. 

```python
mean_salary = df['salary_in_usd'].mean()
print(F"mean_salary is: {mean_salary}")

# Get the unique elements in the 'job_title' column
unique_job_titles= df['job_title'].unique()

average_salary_dict = dict()
for unique_job in unique_job_titles:
    people_with_this_job = df[df['job_title'] == unique_job]
    average_salary_dict[unique_job] = people_with_this_job['salary_in_usd'].mean()

print("Average salary by job:")
pprint.pprint(average_salary_dict)
```

We can instead compute the median salary of each job. 

```python
# Get the unique elements in the 'job_title' column
unique_job_titles= df['job_title'].unique()

average_salary_dict = dict()
for unique_job in unique_job_titles:
    people_with_this_job = df[df['job_title'] == unique_job]
    average_salary_dict[unique_job] = people_with_this_job['salary_in_usd'].median()

print("median salary by job:")
pprint.pprint(average_salary_dict)


```

Delta between media and mean. 

```python
# Get the unique elements in the 'job_title' column
unique_job_titles= df['job_title'].unique()

median_vs_mean_delta = dict()
for unique_job in unique_job_titles:
    people_with_this_job = df[df['job_title'] == unique_job]
    median_vs_mean_delta [unique_job] = abs(people_with_this_job['salary_in_usd'].median() - people_with_this_job['salary_in_usd'].mean())

print("median salary by job:")
pprint.pprint(median_vs_mean_delta )
```

We can always jump to numpy to actually do matrix computations. 

```python
def compute_skewness(data):
    # Step 1: Calculate the mean
    mean = np.mean(data)

    # Step 2: Calculate the standard deviation
    std = np.std(data)

    # Step 3: Subtract the mean from each value
    data_minus_mean = data - mean

    # Step 4: Cube each value
    cubed_values = np.power(data_minus_mean, 3)

    # Step 5: Calculate the mean of the cubed values
    mean_cubed = np.mean(cubed_values)

    # Step 6: Divide the mean of the cubed values by the cube of the standard deviation
    skewness = mean_cubed / (std**3)

    return skewness
```
